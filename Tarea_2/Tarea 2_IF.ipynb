### 2.1. Clase para clasificador (0.5 pt)

Programa una clase `MyMultinomialNB` que en su inicializador reciba el parametro de generalizaci√≥n `alpha`.

```python
class MyMultinomialNB(BaseEstimator, ClassifierMixin):
  def __init__(self, ...):
    ...
```

Para m√°s informacion sobre la construcci√≥n de esta clase puedes revisar [aqu√≠](https://sklearn-template.readthedocs.io/en/latest/user_guide.html#classifier) 

Una llamada de ejemplo para crear un objeto de tu clase ser√≠a:
```python
my_clf = MyMultinomialNB(alpha=1)
``` 
lo que debiera crear un clasificador con par√°metro `alpha` igual a 1.

````python

from sklearn.base import BaseEstimator, ClassifierMixin
import numpy as np

#Naive bayes multinomial classifier
class MyMultinomialNB(BaseEstimator, ClassifierMixin):

    def __init__(self, alpha=1.0):
        self.alpha = alpha
        self.classes = None
        self.n_classes = None
        self.n_features = None
        self.n_samples = None
        self.n_samples_class = None
        self.prior = None
        self.likelihood = None

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.n_classes = len(self.classes)
        self.n_features = X.shape[1]
        self.n_samples = X.shape[0]
        self.n_samples_class = np.zeros(self.n_classes)
        self.prior = np.zeros(self.n_classes)
        self.likelihood = np.zeros((self.n_classes, self.n_features))
        for i in range(self.n_classes):
            X_i = X[y == self.classes[i]]
            self.n_samples_class[i] = X_i.shape[0]
            self.prior[i] = self.n_samples_class[i] / self.n_samples
            self.likelihood[i] = (np.sum(X_i, axis=0) + self.alpha) / (np.sum(X_i) + self.alpha * self.n_features)
        return self
    

    def predict(self, X):
        n_samples = X.shape[0]
        y_pred = np.zeros(n_samples)
        for i in range(n_samples):
            y_pred[i] = self.classes[np.argmax(np.log(self.prior) + np.sum(np.log(self.likelihood) * X[i], axis=1))]
        return y_pred
    
````
